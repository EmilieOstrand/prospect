```python
from probabilit.modeling import Distribution
```

```python
th = Distribution("norm", loc = 50, scale = 10)
ntg = Distribution("norm", loc = 0.6, scale = 0.1)
area = Distribution("norm", loc = 100, scale = 20)
porosity = Distribution("lognorm", s =0.1, scale = 0.1 )

gpv = th * ntg * area * porosity
```

```python
gpv
```

```python
samples = gpv.sample(200, random_state=42)
```

```python
samples
```

```python
import matplotlib.pyplot as plt

plt.hist(samples)
```
